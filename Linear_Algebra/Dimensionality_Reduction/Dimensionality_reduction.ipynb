# What is Dimensionality Reduction

Dimensionality Reduction হচ্ছে একটা technique যেটা আমাদের data set তে  কিছু feature(Dimension) গুলোকে বাদ দিয়ে বাকি feature(Dimension) গুলোকে নিয়ে  নতুন data set বানিয়ে দেয়।

## Dimensionality Reduction কিভাবে কাজ করে?

Dimensionality Reduction technique গুলোর কাজ হচ্ছে  কিছু feature(Dimension) গুলোকে বাদ দিয়ে বাকি feature গুলোকে নিয়ে  নতুন data set বানিয়ে।

## Dimensionality Reduction কেন লাগবে?

 

যদি feature বেশি হয় সেটা process করতেও time লাগে।তাই TIME টা কমানোর জন্য আমরা Dimensionality reduction করতে। এটাতে accuracy
কমলেও time যে problem সেটা আর থাকে না।
