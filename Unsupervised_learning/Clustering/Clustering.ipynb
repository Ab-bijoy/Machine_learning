# What is Clustering ???

### In this technique, which groups the unlabelled dataset.

### A way of grouping the data points into different clusters, consisting of similar data points.
### The objects with possible similarities remain in a group that has less or no similarities with another group.

### --- it is used by Amazon,Netflix in its recommendation system to provide recommendations as per the past search of products,movies respectivity.


# Type of Clustering methods

### Partitioning Clustering



It divides the data into non-hierarchical groups.It is also known as Centroid Based Method.


Example: K-means Clustering Algorithm

 

### Density based Clustering



It connects Highly dense area into clusters, and arbitrarily shaped distributions are fromed as long as dense region can be connected.

Example: DBSCAN, LOF, and Isolation Forest




### Distribution Model Based Clustering



The data is divided based on probability of how a dataset belongs to particular distribution.The grouping is done by assuming some distribution commonly.

Example : Gaussian Distribution

 

### Hierarchical Clustering



It can be used as alternative for Partitioned clustering as there is no requirement of pre-specifying no. of clusters to be created.
Dataset is divided into clusters to create Dendrogram(tree like)

Example: all files and folders on the hard disk are organized in a hierarchy.

 

### Fuzzy Clustering

It is type of soft method(  cluster   cluster   connected) in which data object may belong to more than one cluster. 

Each dataset has a set of membership coefficients which depend on degree of membership to be in a cluster.

Example:
an apple can be red or green (hard clustering), but an apple can also be red AND green (fuzzy clustering). Here, the apple can be red to a certain degree as well as green to a certain degree.
